```python
!pip install mysql-connector-python

import mysql.connector

import mysql.connector
import pandas as pd
import os

# Create folder for CSVs
os.makedirs("csv_results", exist_ok=True)

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="your_username",
    password="your_password",
    database="your_database"
)

# List of queries
queries = {
    "fraud_summary": "SELECT is_fraud, COUNT(*) AS total FROM transactions GROUP BY is_fraud;",
    "amount_summary": "SELECT MIN(amount) AS min_amt, MAX(amount) AS max_amt, AVG(amount) AS avg_amt FROM transactions;",
    "fraud_percentage": "SELECT ROUND(100 * SUM(is_fraud) / COUNT(*), 3) AS fraud_percentage FROM transactions;",
    "top_fraud_transactions": "SELECT * FROM transactions WHERE is_fraud=1 ORDER BY amount DESC LIMIT 10;",
    "transactions_by_hour": """
        SELECT transaction_hour AS hour, COUNT(*) AS total_transactions
        FROM transactions
        GROUP BY transaction_hour
        ORDER BY transaction_hour;
    """,
    "fraud_vs_legit_by_hour": """
        SELECT transaction_hour,
               SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) AS fraud_transactions,
               SUM(CASE WHEN is_fraud=0 THEN 1 ELSE 0 END) AS legit_transactions
        FROM transactions
        GROUP BY transaction_hour
        ORDER BY transaction_hour;
    """,
    "fraud_by_hour_percentage": """
        SELECT transaction_hour,
               SUM(CASE WHEN is_fraud = 1 THEN 1 ELSE 0 END) AS fraud_transactions,
               COUNT(*) AS total_transactions,
               ROUND(SUM(CASE WHEN is_fraud = 1 THEN 1 ELSE 0 END) / COUNT(*) * 100, 2) AS fraud_percentage
        FROM transactions
        GROUP BY transaction_hour
        ORDER BY transaction_hour;
    """,
    "fraud_by_merchant": """
        SELECT merchant_category,
               COUNT(*) AS total_transactions,
               SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) AS fraud_transactions,
               ROUND(SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) / COUNT(*) *100,2) AS fraud_percentage
        FROM transactions
        GROUP BY merchant_category
        ORDER BY fraud_percentage DESC
        LIMIT 10;
    """,
    "fraud_by_device_score": """
        SELECT device_trust_score,
               COUNT(*) AS total_transactions,
               SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) AS fraud_transactions,
               ROUND(SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) / COUNT(*) *100,2) AS fraud_percentage
        FROM transactions
        GROUP BY device_trust_score
        ORDER BY fraud_percentage DESC;
    """,
    "fraud_by_age_group": """
        SELECT CASE 
                   WHEN cardholder_age < 25 THEN '<25'
                   WHEN cardholder_age BETWEEN 25 AND 40 THEN '25-40'
                   WHEN cardholder_age BETWEEN 41 AND 60 THEN '41-60'
                   ELSE '>60'
               END AS age_group,
               COUNT(*) AS total_transactions,
               SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) AS fraud_transactions,
               ROUND(SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) / COUNT(*) *100,2) AS fraud_percentage
        FROM transactions
        GROUP BY age_group
        ORDER BY fraud_percentage DESC;
    """,
    "fraud_by_foreign_category": """
        SELECT foreign_category,
               COUNT(*) AS total_transactions,
               SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) AS fraud_transactions,
               ROUND(SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) / COUNT(*) *100,2) AS fraud_percentage
        FROM transactions
        GROUP BY foreign_category
        ORDER BY fraud_percentage DESC;
    """,
    "fraud_by_velocity": """
        SELECT velocity_last_24h,
               COUNT(*) AS total_transactions,
               SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) AS fraud_transactions,
               ROUND(SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) / COUNT(*) *100,2) AS fraud_percentage
        FROM transactions
        GROUP BY velocity_last_24h
        ORDER BY velocity_last_24h DESC;
    """,
    "fraud_by_location_mismatch": """
        SELECT location_mismathch,
               COUNT(*) AS total_transactions,
               SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) AS fraud_transactions,
               ROUND(SUM(CASE WHEN is_fraud=1 THEN 1 ELSE 0 END) / COUNT(*) *100,2) AS fraud_percentage
        FROM transactions
        GROUP BY location_mismathch
        ORDER BY fraud_percentage DESC;
    """
}

# Execute queries and save to CSV
for name, query in queries.items():
    df = pd.read_sql(query, conn)
    df.to_csv(f"csv_results/{queries_credit_card_analysis}.csv", index=False)
    print(f"Saved csv_results/{queries_credit_card_analysis}.csv")

conn.close()
print("All queries executed and CSVs saved successfully.")


```
